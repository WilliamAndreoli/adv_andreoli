GRÁFICOS DE DISPERSÃO

```python
# Gráficos de dispersão
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregamento da base de dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Gráfico de dispersão considerando volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color = 'red', facecolor = 'yellow', marker = 'o')
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
plt.show
```

```python
# Gráfico de linha considerando o volume e o atributo "girth" - sem marker
#seaborn
sns.lineplot(x=base.Girth, y=base.Volume, color = 'red').set_title('Arvores')
```

```python
# Gráfico de linha considerando o volume e o atributo "girth" - sem marker
#matplotlib
plt.plot(base.Girth, base.Volume, color = 'red')
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão com 'afastamento' dos dados (jilter)
#fit_reg linha de tendência
sns.regplot(x=base.Girth, y=base.Volume, color = 'blue', data = base, fit_reg = False, x_jitter = 0.4, marker='*').set_title('Arvores')
```

GRÁFICO DE DISPERSÃO COM LEGENDA

```python
# Gráfico de dispersão com legenda
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('co2.csv')
base.head()
```

```python
# criando duas variáveis para cada atributo (x = conc e y = uptake)
x = base.conc
y = base.uptake
```

```python
# Retorna os valores únicos do atributo "treatment"
unicos = list(set(base.Treatment))
unicos
```

```python
# Percorre cada tipo de tratamento (chilled e nonchilled) e cria um gráfico de dispersão
for i in range(len(unicos)):
  indice = base.Treatment == unicos[i]
  plt.scatter(x[base.Treatment == unicos[i]], y[base.Treatment == unicos[i]], label = unicos[i])
plt.legend(loc = 'lower right')
```

GRÁFICO DE DISPERSÃO COM SEABORN

```python
# Gráfico de dispersão com seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando a base de dados
base = pd.read_csv('co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
sns.scatterplot(x=base.conc, y=base.uptake, hue=base.Type)
```

```python
# Seleção de registros especificos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=q.conc, y=q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sns.scatterplot(x=m.conc, y=m.uptake).set_title('Mississippi')
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
# Gráfico somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=ch.conc, y=ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
sns.scatterplot(x=nc.conc, y=nc.uptake).set_title('Nonchilled')
plt.tight_layout()
```

GRÁFICOS 3D

```python
# Gráficos 3d
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Carregando a base de dados
base = pd.read_csv('orchard.csv')
base.head()
```

```python
# Criação do gráfico 3D, indicando e atributindo projection - '3d' e passando três atributos
#(decrease, rowpas e colpas)
figure = plt.figure()
eixo = figure.add_subplot(111, projection = '3d')
eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
